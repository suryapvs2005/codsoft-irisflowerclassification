```python
sns.pairplot(df, hue='species')
plt.show()
```

```python
sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 8))

for i, column in enumerate(df.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[column], kde=True, bins=20, color='blue')
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```

```python
# Convert the 'species' column to a categorical data type.
df['species'] = df['species'].astype('category')

# Calculate the correlation matrix only on numerical columns
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12, 8))

for i, column in enumerate(df.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='species', y=column, data=df, palette="Set3")
    plt.title(f'Boxplot of {column} by Species')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

for i, column in enumerate(df.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.violinplot(x='species', y=column, data=df, palette="Set2", inner="quart")
    plt.title(f'Violin Plot of {column} by Species')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('species', axis=1)
y = df['species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression(random_state=0)
lr_model.fit(X_train, y_train)
```

```python
y_lr_pred= lr_model.predict(X_test)
y_lr_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score,  confusion_matrix
lr_cr=classification_report(y_test, y_lr_pred)
print(lr_cr)
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_lr_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix for Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )
knn_model.fit(X_train, y_train)
```

```python
y_knn_pred= knn_model.predict(X_test)
y_knn_pred
```

```python
knn_cr=classification_report(y_test, y_knn_pred)
print(knn_cr)
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_knn_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix for KNN')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model= DecisionTreeClassifier(criterion='entropy', random_state=0)
dt_model.fit(X_train, y_train)
```

```python
y_dt_pred= dt_model.predict(X_test)
y_dt_pred
```

```python
dt_cr=classification_report(y_test, y_dt_pred)
print(dt_cr)
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_dt_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix for Decision Tree')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
```

```python
y_rf_pred= rf_model.predict(X_test)
y_rf_pred
```

```python
rf_cr=classification_report(y_test, y_rf_pred)
print(rf_cr)
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_rf_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix for Random Forest Classifier')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train, y_train)
```

```python
y_svc_pred= svc_model.predict(X_test)
y_svc_pred
```

```python
svc_cr=classification_report(y_test, y_svc_pred)
print(svc_cr)
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_svc_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix for SVC')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
models = ['Logistic Regression', 'Decision Tree','Random Forest', 'Support Vector Machine (SVM)', 'K Nearest Neighbors']
accuracy = [ 0.93, 0.93, 0.96, 0.96, 0.96]

bar_width = 0.35

r1 = np.arange(len(accuracy))


plt.bar(r1, accuracy, color=['b', 'g', 'yellow', 'r', 'orange', 'grey'], width=bar_width, edgecolor='grey')

plt.xlabel('Model', fontweight='bold')
plt.ylabel('Accuracy Score')
plt.xticks([r1 + bar_width/15 for r1 in range(len(accuracy))], models, rotation=90)
plt.title('Comparison of Accuracy Scores between different Models')

plt.legend()

plt.show()
```
